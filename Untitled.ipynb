```python
import sklearn
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import LinearSVC
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
import scipy.sparse
import numpy as np
import pandas as pd
import random
import seaborn as sns
import matplotlib as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# get adult dataset into pandas df and shuffle
data_adult = pd.read_csv("./datasets/adult.data")
data_adult.dropna()
adult_dataset = pd.get_dummies(data_adult,drop_first=True).values
np.random.shuffle(adult_dataset)
```

```python
data_adult.head()
```

Adult dataset

Using Logistic Regression:

```python
#classifier tuning
C_list = [.001,.01,.1,1,10]
parameters = {'C':C_list}
```

```python
# features of dataset
data_len = len(adult_dataset)

train_2080_adult = adult_dataset[:int(.2*data_len):,0:-1]
train_5050_adult = adult_dataset[:int(.5*data_len):,0:-1]
train_8020_adult = adult_dataset[:int(.8*data_len):,0:-1]

# classification
train_2080_adult_y = adult_dataset[:int(.2*data_len),-1]
train_5050_adult_y = adult_dataset[:int(.5*data_len),-1]
train_8020_adult_y = adult_dataset[:int(.8*data_len),-1]
```

```python
log_reg = LogisticRegression(solver = "liblinear", class_weight = "balanced")
C_list = [.001,.01,.1,1,10]
parameters = {'C':C_list}

clf = GridSearchCV(log_reg, parameters, cv = 3, return_train_score=True)

clf.fit(train_2080_adult,train_2080_adult_y)
print("Logistic regression test 2080: ", clf.cv_results_.get('mean_test_score'))
print("Logistic regression train 2080: ",clf.cv_results_.get('mean_train_score'))
print("accuracy score 2080:", accuracy_score(train_2080_adult_y, clf.predict(train_2080_adult)))
print("f1 score 2080:",f1_score(train_2080_adult_y, clf.predict(train_2080_adult)),'\n')

clf.fit(train_5050_adult,train_5050_adult_y)
print("Logistic regression test 5050: ", clf.cv_results_.get('mean_test_score'))
print("Logistic regression train 5050: ",clf.cv_results_.get('mean_train_score'))
print("accuracy score 5050:", accuracy_score(train_5050_adult_y, clf.predict(train_5050_adult)))
print("f1 score 5050:",f1_score(train_5050_adult_y, clf.predict(train_5050_adult)),'\n')

clf.fit(train_8020_adult,train_8020_adult_y)
print("Logistic regression test 8020: ", clf.cv_results_.get('mean_test_score'))
print("Logistic regression train 8020: ",clf.cv_results_.get('mean_train_score'))
print("accuracy score 8020:", accuracy_score(train_8020_adult_y, clf.predict(train_8020_adult)))
print("f1 score 8020:",f1_score(train_8020_adult_y, clf.predict(train_8020_adult)),'\n')
```
